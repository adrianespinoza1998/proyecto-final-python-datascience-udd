# Inspección de datos

- game_id (int): Identificador único de la partida.
- rated (bool): Si la partida fue clasificada o no (True, False).
- turns (int): Número de turnos en la partida. 
- victory_status (string): Estado de la partida (Out of Time, Resign, Mate, Draw).
- winner (string): Ganador de la partida (White, Black, Draw).
- time_increment (string): Duración inicial de la partida + incremento de tiempo en segundos, Ej: 15+2 => Partida con duración base de 15 minutos + 2 segundos de adición por movimiento.
- white_id (string): ID del jugador blanco.
- white_rating (int): Rating del jugador blanco.
- black_id (string): ID del jugador negro.
- black_rating (int): Rating del jugador negro.
- moves (string): Todos los movimientos realizados en la partida.
- opening_code (string): Código ECO usado para categorizar la apertura.
- opening_moves (int): Número de movimientos realizados en la apertura. 
- opening_fullname (string): Nombre completo de la apertura realizada.
- opening_shortname (string): Nombre abreviado de la apertura.
- opening_response (string): Respuesta por parte de las negras a la apertura (Refused, Accepted, Declined)
- opening_variation (string): Variación de la apertura.
